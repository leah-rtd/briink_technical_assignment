# [Metrics Documentation](https://wikirate.org/Metrics_Introduction)

What are these metrics ?
 - a Metric consists of a Question, it contains an About section explaining the question being asked, and a Methodology section describing how the answers are researched.   

How and who can add them to the platform ?
 - Answers to standard research metric questions are entered either individually via research projects or in bulk by imports. All researched answers must cite a public source and can be verified (checked) by other users.   


Different kinds of metrics ? 
 - calculated metrics: refers to the metric type being based on a calculated value, and can come in the form of a Formula, Score, Rating, Descendant, and Network-Aware,
 - score metrics: assign a maximum of 10 points to the different answer options of another metric, which can form the basis for any kind of rating
 - ratings: A Rating is a 0-10 rating computed as a weighted average of other 0-10 answers. To create a Rating, a number of Score or other Rating metrics are selected and assigned weightings.
 - descendants: inherits answers from one or more ancestor metrics. Descendant metrics are used to pool the data from a set of similar input metrics,
 - formulas: derive answers dynamically from the answers to other metrics. Formula metrics give you the flexibility to create formulae directly, while other calculated metric types have an interface for creating a specific type of formula,
 - relations metrics: describe the relationship between two companies,
 - network-aware metrics: dynamically measure a company’s performance based on the performance of related companies

## Retrieving Metric Question & Custom_id

JSON with 3 fields: `input`, `reference_output`, `structured_data`

`input`:
- question : str
- custom_id :str
- data_type : "Metric" (equal to value_type=Number)
- company : str
- file_metas : list of dicts

`reference_output`:
- answer : str
- source_documents : list of dicts (equal to file_metas and maybe page_number)

`structured_data`:
- list of dict with `unit` : str, `value` : float, `time_period` : str

# Refactoring into methods

```python
def get_metric_info(metric):
    url =  f"https://wikirate.org/~{metric}"


    params = {"api_key": api_key,
            "format": "json"
    }

    response_metric = requests.get(url, params=params)
    question = ""
    custom_id = ""
    data_type = ""
    if response_metric.status_code == 200:
        response_metric = response_metric.json()
        question = response_metric.get("question", {}).get("content", "")
        custom_id = response_metric.get("name", "")
        if response_metric.get("value_type", {}).get("content") == "Number":
            data_type = "Metric"
        else:
            data_type = ""

    return {"question": question,
            "custom_id": custom_id,
            "data_type": data_type
    }
```

```python
get_metric_info(m_nb_employees)
```

```python
def get_company_info(metric, company, year):
    url = f"https://wikirate.org/~{metric}+~{company}+{year}"

    params = {"api_key": api_key,
            "format": "json"
    }

    response = requests.get(url, params=params)

    company = ""
    file_metas = []
    answer = ""
    structured_data = []
    if response.status_code == 200:
        response = response.json()
        company = response.get("company", "")
        sources = response.get("sources", [])
        for source in sources:
            file_metas.append(
                {"file_name": source.get("title", ""),
                "file_url": source.get("url", "")}
            )
        answer = response.get("value", "") + " " + response.get("unit", "")
        structured_data = [
            {"unit": response.get("unit", ""),
            "value": float(response.get("value", "")),
            "time_period": response.get("year")}
        ]

    return {"company": company,
            "file_metas": file_metas,
            "answer": answer,
            "structured_data": structured_data
    }
```

```python
get_company_info(m_nb_employees, comp_hm, 2016)
```

```python
def get_all_company_metric(metric, companies, year):
    metric_info = get_metric_info(metric)

    result = []
    for company in companies:
        company_info = get_company_info(metric, company, year)
        result.append(
            {
                "input": {
                    "question": metric_info["question"],
                    "custom_id": metric_info["custom_id"],
                    "data_type": metric_info["data_type"],
                    "company": company_info["company"],
                    "file_metas": company_info["file_metas"]
                },
                "reference_output": {
                    "answer": company_info["answer"],
                    "source_documents": company_info["file_metas"]
                },
                "structured_data": company_info["structured_data"]
            }
        )

    return result
```
